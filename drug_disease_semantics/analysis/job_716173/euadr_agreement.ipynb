```python
def create_match(gold_std_answer):
    def match(identifier, crowd_choice):
        return int(crowd_choice == gold_std_answer[identifier])
    
    return match
```

```python
match_obj = create_match(euadr_answer)
```

```python
results = aggregate_results("pub_rel_id", "broad_rel_type", raw_data, match_obj)
```

```python
def determine_matches(results):
    matches = []
    mismatches = []
    for rel_id, group in results.groupby("pub_rel_id"):
        if group.iloc[0]["match"] == 1:
            matches.append(rel_id)
        else:
            mismatches.append(rel_id)
            
    return (matches, mismatches)
```

```python
matches, mismatches = determine_matches(results)
```

```python
def broad_reltype_summary(settings):
    raw_data = filter_data(settings)

    true_relation_type = get_euadr_relation_type(settings)

    match_obj = create_match(true_relation_type)

    results = aggregate_results("pub_rel_id", "broad_rel_type", raw_data, match_obj)

#--------------------------------------------------------------------
    ans = pd.Series([0] * 4)
    for unit_id, group in results.groupby(settings["id_column"]):
        temp = list(group["match"])
        ans = ans.add(pd.Series(temp), fill_value = 0)

    num_work_units = len(results[settings["id_column"]].unique())

    print "Distribution of matches with gold standard based on response rank:\n{0}".format(ans)
    print
    print ("Total number of work units for this subset of data: {0}".
        format(num_work_units))
    print
    print "Number of correct responses with respect to the rank of confidence score:"
    print ("{0}/{1} ({2}%) work units had the correct answer in the top ranked position".
           format(ans[0], num_work_units, ans[0]/num_work_units * 100))

    graph = ans.plot(kind = "bar")
    graph.set_xlabel("Rank of the correct choice according to the crowd")
    graph.set_ylabel("Number of work units where the crowd chosen top answer\n"\
        "matched that of the gold standard")
```

```python
broad_reltype_summary(settings)
```

```python
orig_problem = get_orig_problem("pub_rel_id", raw_data)
```

```python
def sort_by_conf(work_units):
    """
    Sort the work units in decreasing order of agreement
    for the top chosen crowd answer.
    """
    temp = []
    for rel_id in work_units:
        group = results.query("pub_rel_id == '{0}'".format(rel_id))
        temp.append((rel_id, group.iloc[0]["percent_agree"]))
        
    temp = sorted(temp, key = lambda x: x[1], reverse = True)
    return [val[0] for val in temp]
```

```python
matches = sort_by_conf(matches)
```

```python
mismatches = sort_by_conf(mismatches)
```

```python
def graph_confidence(work_units):
    conf = []
    
    for rel_id in work_units:
        group = results.query("pub_rel_id == '{0}'".format(rel_id))
        
        conf.append(group.iloc[0]["percent_agree"])
        
    graph = pd.Series(sorted(conf, reverse = True)).plot()
    
    graph.set_ylim((0, 1.1))
```
