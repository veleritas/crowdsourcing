```python
def create_match(gold_std_answer):
    def match(identifier, crowd_choice):
        return int(crowd_choice == gold_std_answer[identifier])
    
    return match
```

```python
match_obj = create_match(euadr_answer)
```

```python
results = aggregate_results("pub_rel_id", "broad_rel_type", raw_data, match_obj)
```

Look at the results:

```python
def determine_matches(results):
    matches = []
    mismatches = []
    for rel_id, group in results.groupby("pub_rel_id"):
        if group.iloc[0]["match"] == 1:
            matches.append(rel_id)
        else:
            mismatches.append(rel_id)
            
    return (matches, mismatches)
```

```python
matches, mismatches = determine_matches(results)
```

```python
def deep_sem_summary(work_units):
    temp = []
    for rel_id in work_units:
        subset = results.query("pub_rel_id == '{0}'".format(rel_id))
        top_choice = subset.iloc[0]["broad_rel_type"]
        
        if top_choice not in ["false", "negative"]:
            raw_subset = raw_data.query("pub_rel_id == '{0}'".format(rel_id))

            top_subset = raw_subset.query("broad_rel_type == '{0}'".format(top_choice))

            ans = aggregate_votes("deep_semantics", top_subset)
            
            total_vote_score = sum(ans["conf_score"])
            percent_agree = ans["conf_score"] / total_vote_score
            
            ans["percent_agree"] = percent_agree
            ans.insert(0, "pub_rel_id", rel_id)
            
            ans = ans.sort(["conf_score"], ascending = False)
            
            temp.append(ans)
            
    return pd.concat(temp)
```

```python
deep_sem_res = deep_sem_summary(matches)
```
