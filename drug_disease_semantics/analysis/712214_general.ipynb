```python
def matches(gold_ans, crowd_ans):
    mapping = {
        "definite_relation": "positive",
        "uncertain_relation": "speculative",
        "inferred_relation": "false",
        "no_relation": "negative",
        "is_cooccurrence": "false"
    }
    
    return int(gold_ans == mapping[crowd_ans])

def determine_broad_reltype_results(true_relation_type, data):
    """
    Calculates the agreement percentage and determines whether the
    top crowd response matches the gold standard for all work units.
    """
    result = []
    for unit_id, group in data.groupby("_unit_id"):
        ans = aggregate_votes("sentence_claim", group)

        total_vote_score = sum(ans["conf_score"])
        percent_agreement = ans["conf_score"] / total_vote_score
        
        match = [matches(true_relation_type[unit_id], val) for val in ans["sentence_claim"]]

        ans.insert(len(ans.columns), "percent_agreement", percent_agreement)
        ans.insert(len(ans.columns), "match", match)
        ans.insert(0, "unit_id", unit_id)

        ans = ans.sort(["conf_score"], ascending = False)

        result.append(ans)

    return pd.concat(result)
```

```python
def broad_reltype_summary(settings):
    data = filter_data(settings)
    true_relation_type = get_euadr_relation_type(settings)
    results = determine_broad_reltype_results(true_relation_type, data)

#    if "show_raw" in settings and settings["show_raw"]:
#        print results

    ans = pd.Series([0] * 4)
    for unit_id, group in results.groupby("unit_id"):
        temp = list(group["match"]) # reset the indicies
        ans = ans.add(pd.Series(temp), fill_value = 0)

    num_work_units = len(results["unit_id"].unique())

    print "Distribution of matches with gold standard based on response rank:\n{0}".format(ans)
    print
    print ("Total number of work units for this subset of data: {0}".
        format(num_work_units))
    print
    print "Number of correct responses with respect to the rank of confidence score:"
    print ("{0}/{1} ({2}%) work units had the correct answer in the top ranked position".
           format(ans[0], num_work_units, ans[0]/num_work_units * 100))

    graph = ans.plot(kind = "bar")
    graph.set_xlabel("Rank of the correct choice according to the crowd")
    graph.set_ylabel("Number of work units where the crowd chosen top answer\n"\
        "matched that of the gold standard")
```

```python
# get crowd ans:
crowd_broad = dict()
for unit_id, group in results.groupby("unit_id"):
    top_choice = group.iloc[0]["sentence_claim"]
    
    crowd_broad[unit_id] = top_choice
```

```python
def compare_responses(group_a, group_b):
    """
    Broad semantics.
    Compares to find the unique work ids that groups a and b agreed upon or disagreed upon.
    """
    agree = set()
    disagree = set()
    
    # there is no way to evaluate work units which only one group worked on
    common_units = set(group_a.keys()) & set(group_b.keys())
    print "Common number of work units:", len(common_units)
    for unit_id in common_units:
        if matches(group_a[unit_id], group_b[unit_id]): # gold then crowd
            agree.add(unit_id)
        else:
            disagree.add(unit_id)
            
    return (agree, disagree)
```

```python
agree, disagree = compare_responses(euadr_answer, crowd_broad)

print "Num agree:", len(agree)
print "Num disagree:", len(disagree)
```
