Convert to a common language for the broad relationship types

```python
def convert(resp):
    ans = {"PA": "positive", "FA": "false", "SA": "speculative", "NA": "negative"}
    assert resp in ans
    return ans[resp]
```

Read in our data

```python
def read_data(fname, loc, want_golden = False):
    """
    Reads the raw data from CrowdFlower at the specified location.
    Returns the true relation type for each work unit, as well as a list
    of the human judgements for that work unit.
    """
    true_relation_type = dict()
    human_response = defaultdict(lambda: defaultdict(tuple))

    for line in islice(read_file(fname, loc), 1, None):
        vals = line.split('\t')
        assert len(vals) == 36

        unit_id = vals[0]
        is_golden = vals[1] == "true"
        trust_score = float(vals[6])
        worker_id = vals[7]
        correlation_type = vals[9]
        true_ans = convert(vals[34])

        choices = ["positive", "negative", "false", "speculative"]
        # check that data is ok
        assert true_ans in choices
        assert correlation_type in choices
        assert 0 <= trust_score <= 1

        if is_golden and not want_golden:
            continue
        
        
        
        true_relation_type[unit_id] = true_ans
        # worker answer for this particular question
        human_response[unit_id][worker_id] = (correlation_type, trust_score)
        
    return (true_relation_type, human_response)
```

```python
def correct_link_type(target, sub_type, obj_type):
#   drug/disease
#   gene/disease
#   variant/disease
#   drug/gene
#   variant/drug

    return target in [sub_type + "/" + obj_type, obj_type + "/" + sub_type]

def read_data2(fname, loc, **settings):
    """
    Reads raw CrowdFlower output, and returns the information
    based on the settings.

    Possible settings:
    data_subset = ["gold", "normal", "both"]
    min_accuracy = a floating point number from 0 to 1
    include_unfinished = a boolean (right now is true by default)
    link_type = ["gene drug", "drug disease", "gene disease"]
    """
    true_relation_type = dict()
    human_judgement = defaultdict(lambda: defaultdict(tuple))
    for line in islice(read_file(fname, loc), 1, None): # skip header
        vals = line.split('\t')
        assert len(vals) == 42

        unit_id = vals[0]
        is_golden = (vals[2] == "true")
        accuracy = float(vals[8])
        worker_id = vals[9]
        broad_relationship = vals[15]

        sub_type = vals[39]
        obj_type = vals[34]
        true_broad_relationship = convert(vals[40])

        choices = ["positive", "negative", "speculative", "false"]

#       check that the data are ok
        assert true_broad_relationship in choices
        assert broad_relationship in choices
        assert 0 <= accuracy <= 1

#       filtering steps depending on the settings

        if is_golden and settings["data_subset"] == "normal":
            continue

        if not is_golden and settings["data_subset"] == "gold":
            continue

        if accuracy < settings["min_accuracy"]:
            continue

        if not correct_link_type(settings["link_type"], sub_type, obj_type):
            continue

        true_relation_type[unit_id] = true_broad_relationship
        human_judgement[unit_id][worker_id] = (broad_relationship, accuracy)

    return (true_relation_type, human_judgement)
```

```python
def determine_matches(true_relation_type, human_response, accuracy_threshold):
    """
    Returns a list of each work unit, the percentage of the total confidence score
    of the top choice, and whether that choice matched the gold standard.
    """
    # answer[unit_id] = (top human chosen relation, total confidence score for that choice,
    # and the percentage agreement for that choice, the gold standard answer, and whether
    # the human response matched the gold standard)
    answer = dict()
    for unit_id, responses in human_response.items():
        score = defaultdict(float)
        for worker_id, response in responses.items():
            if response[1] >= accuracy_threshold:
                score[response[0]] += response[1]

        ans = [(rel_type, conf_score) for rel_type, conf_score in score.items()]
        ans = sorted(ans, key = lambda x: x[1], reverse = True)

        if len(ans) > 1: # check that there are no ties
            assert ans[0][1] > ans[1][1]
            
        # find the % out of the total
        agreement = ans[0][1] / sum([conf_score for rel_type, conf_score in ans])
        match = int(true_relation_type[unit_id] == ans[0][0])
        answer[unit_id] = (ans[0][0], ans[0][1], agreement, true_relation_type[unit_id], match)
        
    return answer
```

```python
loc = "/home/toby/crowdsourcing/extract_relationships/results/analysis"
fname = "705001_full_with_untrusted_and_unfinished.csv"

settings = {
    "data_subset": "both",
    "min_accuracy": 0.7,
    "link_type": "drug/disease"
}


outname = "lol.txt"
accuracy_threshold = 0.7

true_relation_type, human_response = read_data2(fname, loc, **settings)
print len(human_response)
match_results = determine_matches(true_relation_type, human_response, accuracy_threshold)


# write to file so R can use the data
header = ["unit_id", "best_human_relation_type", "best_human_relation_conf_score",
          "agreement_percentage", "true_relation_type", "match"]

with open(outname, "w") as out:
    out.write("{0}\n".format("\t".join(header)))
    for unit_id, data in match_results.items():
        data = map(str, data)
        out.write("{0}\t{1}\n".format(unit_id, "\t".join(data)))


```

```python
%%R

# draws an ROC curve with the given data, and prints the AUC for that curve
draw_ROC <- function(score, result, title)
{
    pred <- prediction(score, result)
    perf <- performance(pred, measure = "tpr", x.measure = "fpr")
    
    plot(perf, colorize = TRUE, colorkey = TRUE, lwd = 5,
        main = title)
    
    temp <- performance(pred, "auc")
    auc <- slot(temp, "y.values")[[1]]
    
    print(paste0("AUC: ", auc))
}
```
