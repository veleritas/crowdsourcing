```python
def get_crowd_choices(gold_std, settings):
    """
    Returns a dict with the crowd's top answer for each work unit.
    """
    crowd_data = filter_data(settings)
    crowd_results = determine_broad_relationship_matches(gold_std, crowd_data)

    ans = dict()
    for unit_id, group in crowd_results.groupby("unit_id"):
        resp = group["relation_type"].iloc[0]
        ans[unit_id] = resp
        
    return ans
```

```python
def get_toby_answers():
    loc = "/home/toby/crowdsourcing/extract_relationships/results/analysis/"
    fname = "toby_annotations_for_cf_705001.txt"

    data = pd.read_csv(loc + fname, sep = '\t')
    data = data.fillna("empty")
    
    toby_relations = dict()
    deep_sem = dict()
    for unit_id, group in data.groupby("unit_id"):
        assert len(group["toby_association_type"].unique()) == 1
        toby_relations[unit_id] = group["toby_association_type"].iloc[0]
        deep_sem[unit_id] = group["toby_deep_semantics"].iloc[0]
        
    return (toby_relations, deep_sem)
```

```python
toby_relations, toby_deep_sem = get_toby_answers()
```

```python
def sem_name(sub_type, obj_type):
    if sub_type == "disease":
        assert obj_type == "drug"
        return "drug_disease_semantics"
    
    if sub_type == "gene or protein":
        if obj_type == "disease":
            return "gene_disease_semantics"
        elif obj_type == "drug":
            return "gene_drug_semantics"
        
    assert sub_type == "gene variant"
    if obj_type == "disease":
        return "variant_disease_semantics"
    elif obj_type == "drug":
        return "variant_drug_semantics"
        
        
    raise Exception("failure")
```

```python
def aggregate_deep_sem_votes(deep_sem_column, data):
    """
    Aggregate the deep semantic votes together, based on conf score.
    """
    temp = []
    for deep_sem_relationship, group in data.groupby(deep_sem_column):
        conf_score = sum(group["_trust"])
        num_votes = len(group)
        
        temp.append([deep_sem_relationship, conf_score, num_votes])
        
    return pd.DataFrame(temp, columns = ["deep_semantic_choice", "conf_score", "num_votes"])
```

```python
# for a single work unit, and the top choice
# figure out what the voting for deep semantics looks like
def aggregate_deep_semantics(unit_id, top_crowd_broad_choice, data):
    if top_crowd_broad_choice not in ["positive", "speculative"]:
        return "none"
    
    subset = data.query("_unit_id == {0}".format(unit_id))
    subset = subset.query("correlation_type == '{0}'".format(top_crowd_broad_choice))
    
    assert len(subset["subject_type"].unique()) == 1
    assert len(subset["object_type"].unique()) == 1    
    
    sub_type = subset["subject_type"].iloc[0]
    obj_type = subset["object_type"].iloc[0]
    
    
    col_name = sem_name(sub_type, obj_type)

    res = aggregate_deep_sem_votes(col_name, subset)
    res = res.sort(["conf_score"], ascending = False)
    return res
```

```python
import re
def remove_tags(text):
    text = text.replace("&quot;", "")
    text = text.replace("<div class=\"formatted_sentence\" id=\"\" style=\"display: \">", "")
    text = text.replace("</span>", "")
    text = text.replace("<span class=\"subject_text\" id=\"\" style=\"display: \">", "")
    text = text.replace("<span class=\"object_text\" id=\"\" style=\"display: \">", "")
    text = text.replace("</div>", "")
    return text

def get_original_problem(unit_id):
    """
    Returns the subject text, object text, and sentence for this work unit id.
    """
    loc = "/home/toby/crowdsourcing/extract_relationships/results/analysis/"
    fname = "705001_full_with_untrusted_and_unfinished.csv"
    
    raw_data = pd.read_csv(loc + fname, sep = "\t")
    
    subset = raw_data[raw_data["_unit_id"] == unit_id]
    
    sentence = subset["formatted_sentence"].iloc[0]
    sub_text = subset["subject_text"].iloc[0]
    obj_text = subset["object_text"].iloc[0]
    
    sentence = remove_tags(sentence)
    sub_text = remove_tags(sub_text)
    obj_text = remove_tags(obj_text)
    
    sentence = sentence.replace(sub_text, "[[{0}]]".format(sub_text))
    sentence = sentence.replace(obj_text, "[[{0}]]".format(obj_text))
    
    return (sub_text, obj_text, sentence)
```
