### Analysis of Crowdflower job #705001: effect of treating positive and speculative associations as the same thing

```python
import matplotlib.pyplot as plt

%matplotlib inline
%pylab inline
```

```python
from src.summary import data_summary
```

The gold standard differentiates between "positive" and "speculative" assocations. However, the differences can be subtle, and therefore it might be useful to treat these two options as the same thing.

Let's examine the effect of combining the two assocations into one association type.

Let's first look at how the crowd did on the normal data set, where each worker had to have a minimum accuracy score of 70%.

```python
settings = {
    "data_subset": "normal",
    "min_accuracy": 0.7,
    "max_accuracy": 1.0,
    "link_type": "all",
    "loc": "/home/toby/crowdsourcing/extract_relationships/results/analysis",
    "fname": "705001_full_with_untrusted_and_unfinished.csv",
    "show_raw": False,
    "combine_pos_and_spec": False
}

data_summary(settings)
```

So the crowd got the right answer 63.26% of the time, which is ok, but not great. Ideally we would like to see something like 80% to 90% agreement with the gold standard. The signal captured by the top choice is 31/35 = 88.57%.

If we treat positive and speculative assocations as the same thing, then we see an improvement in performance:

```python
settings = {
    "data_subset": "normal",
    "min_accuracy": 0.7,
    "max_accuracy": 1.0,
    "link_type": "all",
    "loc": "/home/toby/crowdsourcing/extract_relationships/results/analysis",
    "fname": "705001_full_with_untrusted_and_unfinished.csv",
    "show_raw": False,
    "combine_pos_and_spec": True
}

data_summary(settings)
```

The crowd now gets 69.38% of the questions correct, compared to the 63.26% when we did not combine the positive associations with the speculative ones.

It's also a good sign that the crowd's second choice does not increase in the number of correct answers it gathers. The signal captured by the top choice here is 34/39 = 87.17%. It has stayed more of less constant. However, to improve the responses, we would like to make sure that the correct answer is not the second choice, but the first choice.

From this we see that when we combine the positive and speculative answers together, the crowd's overall performance improves, without diminishing the amount of signal captured by the top choice.
